## Who is the charterer that brought most gasoline into Mexico?

```python
gasoline_imports = df.query("product_category == 'Finished Gasoline'")
gasoline_imports.primary_vessel_charterer.value_counts()
```

## What amounts did each charterer brought?

```python
#df.columns = df.columns.str.lower().str.replace(" ", "_")
#df=df.rename(columns = {'two':'new_name'})
gasoline_imports = gasoline_imports.rename(columns= {'quantity_(bbl)':'quantity_bbl'})
#gasoline_imports.head()
gasoline_imports.groupby(by='primary_vessel_charterer').quantity_bbl.sum().sort_values(ascending = False)
```
